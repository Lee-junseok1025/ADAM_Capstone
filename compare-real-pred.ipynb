```python
import pandas as pd
import numpy as np
import os
import FinanceDataReader as fdr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
KOSPI200_pred = pd.read_csv('../input/indices/week1_KOSPI200.csv')
KOSDAQ_pred = pd.read_csv('../input/indices/week1_KOSDAQ.csv')
NASDAQ_pred = pd.read_csv('../input/indices/week1_NASDAQ.csv')
SNP500_pred = pd.read_csv('../input/indices/week1_SNP500.csv')
DOWJONES_pred = pd.read_csv('../input/indices/week1_US30.csv')
```

```python
stock_name = ['KOSPI200','SNP500','NASDAQ','KOSDAQ','DOWJONES']
stock_code = ['KS200','US500','IXIC','KQ11','DJI']
```

```python
for i in range(len(stock_code)):
    start_date = '2011-01-01'
    end_date = '2021-11-19'
    stock_name[i] = fdr.DataReader(stock_code[i],start=start_date,end=end_date)[['Close']]
```

```python
KOSPI200 = stock_name[0]
SNP500 = stock_name[1]
NASDAQ = stock_name[2]
KOSDAQ = stock_name[3]
US30 = stock_name[4]
```

```python
for i in range(len(stock_code)):
    start_date = '2021-11-01'
    end_date = '2021-11-26'
    stock_code[i] = fdr.DataReader(stock_code[i],start=start_date,end=end_date)[['Close']]
    
```

```python
real_KOSPI200 = stock_code[0]
real_SNP500 = stock_code[1]
real_NASDAQ = stock_code[2]
real_KOSDAQ = stock_code[3]
real_US30 = stock_code[4]
```

```python
def make_datetime(pred):
    pred = pred['pred_Close_Indices'].to_numpy()
    pred_start = pd.to_datetime('2021-11-01')
    pred_end = pd.to_datetime('2021-11-26')
    name_pred = pd.DataFrame()
    name_pred.index = pd.date_range(pred_start,pred_end,freq='B')
    name_pred['prediction'] = pred
    return name_pred
```

```python
KOSPI200_prediction = make_datetime(KOSPI200_pred)
SNP500_prediction = make_datetime(SNP500_pred)
NASDAQ_prediction = make_datetime(NASDAQ_pred)
KOSDAQ_prediction = make_datetime(KOSDAQ_pred)
DOWJONES_prediction = make_datetime(DOWJONES_pred)
```

```python
def compare_real_pred(stock_name,real,pred):
    plt.figure(figsize=(25,13))
    plt.rc('font',size=20)
    plt.title(stock_name)
    plt.plot(pred,ls='-.',label='prediction_Indices',color='black')
    plt.plot(real,label='real_Indices',color='red')
    plt.legend(loc='upper right')
    plt.grid()
    return plt.show()
```

```python
compare_real_pred('KOSPI200',real_KOSPI200,KOSPI200_prediction)
```

```python
compare_real_pred('S&P500',real_SNP500,SNP500_prediction)
```

```python
compare_real_pred('NASDAQ',real_NASDAQ,NASDAQ_prediction)
```

```python
compare_real_pred('KOSDAQ',real_KOSDAQ,KOSDAQ_prediction)
```

```python
compare_real_pred('DOWJONES',real_US30,DOWJONES_prediction)
```

```python
def plot_real_n_pred(first_data,stock_name,real_data,ensemble_data):
    train = first_data[:-20]
    test = first_data[-20:]
    plt.figure(figsize=(20,10))
    plt.title(f'{stock_name}_prediction')
    plt.xlabel('Date',fontsize=12)
    plt.ylabel('Indices',fontsize=12)
    plt.plot(train,label='train_close',color='blue')
    plt.plot(test,label='test_close',color='yellow')
    plt.plot(real_data,label='Real_Close',color='green')
    plt.plot(ensemble_data,label='pred_close',color='black')
    plt.legend()
    plt.grid(True)
    return plt.show()
    
```

```python
plot_real_n_pred(KOSPI200[2200:],'KOSPI200',real_KOSPI200,KOSPI200_prediction)
```
